## 1) 2D Array - DS

**Task** 

Calculate the hourglass sum for every hourglass in A, then print the maximum hourglass sum.

**Input Format**

There are  lines of input, where each line contains  space-separated integers describing 2D Array ; every value in  will be in the inclusive range of $-9$ to $9$ .


**Output Format**

Print the largest (maximum) hourglass sum found in A.

```python
#!/bin/python3

import math
import os
import random
import re
import sys

max_sum = -float('inf')
# Complete the hourglassSum function below.

def hourglassSum(arr, max_sum):
    results = [max_sum]
    for row in range(1,5):
        for col in range(1,5):
            
            sum = 0 #initialize sum
            sum += arr[row - 1][col - 1] #above left element
            sum += arr[row - 1][col] # directly above
            sum += arr[row - 1][col + 1] # above right
            sum += arr[row][col] # element itself
            sum += arr[row + 1][col - 1]#below left
            sum += arr[row + 1][col] # directly below
            sum += arr[row + 1][col + 1] # below right
            
            results.append(sum)
            
    max_sum = max(results)
            
    return max_sum


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')
    
    arr = []

    for _ in range(6):
        arr.append(list(map(int, input().rstrip().split())))

    result = hourglassSum(arr, max_sum)

    fptr.write(str(result) + '\n')

    fptr.close()
```

## 2) Left Rotation

**Task**

A left rotation operation on an array shifts each of the array's elements  unit to the left. For example, if $2$ left rotations are performed on array $[1,2,3,4,5]$ , then the array would become $[3,4,5,1,2]$.

Given an array $a$ of $n$  integers and a number, $d$ , perform $d$ left rotations on the array. Return the updated array to be printed as a single line of space-separated integers.

```python
A = [2,4,3,5]

# Right (Cyclic) Rotation

print(A[-2:] + A[:-2])

# Left Rotation

print(A[2:] + A[:2])
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the rotLeft function below.
# Here I used slicing, but there are multiple possible solutions
# One of them is provided below

def rotLeft(a, d):
    
    return a[d:] + a[:d]
    

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    nd = input().split()

    n = int(nd[0])

    d = int(nd[1])

    a = list(map(int, input().rstrip().split()))

    result = rotLeft(a, d)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```

```python
# Rotate with Recursion

def rotLeftOnce(a):
    
    tmp = a[0] #save 1st element
    
    for j in range(len(a) - 1):
        
        a[j] = a[j+1] # shift all elements to the left one-by-one
    
    a[j + 1] = tmp #rewrite last element with the save one (1st one)
    
def rotLeft(a, d):
    
    assert (d > 0), "n should be non-negative integer"
    
    for _ in range(d):
        
        rotLeftOnce(a)
        
    print(a)
    
    
rotLeft(A,2)        
```

## 3) Minimum Swaps 2

With given an unordered array consisting of consecutive integers where you are allowed to swap any two elements. You need to find the minimum number of swaps required to sort the array in ascending order.

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the minimumSwaps function below.

def minimumSwaps(arr):   
    dict_map = {}
    
    for j in range(len(arr)):
        dict_map[arr[j]] = j
        
    sorted_arr = sorted(arr)
    swaps = 0
    
    for j in range(len(arr)):
        if arr[j] != sorted_arr[j]:
            swaps += 1
            
            #Get index of sorted current value
            to_swap_ind = dict_map[sorted_arr[j]]
            
            # Update map[**CurrentValue**]=map[**SortedCurrentValue**] 
            # (map[**SortedCurrentValue**] does not need to be updated as we are not going to use it any more)
            dict_map[arr[j]] = dict_map[sorted_arr[j]]
            
            # In input list swap **CurrentValue** with **SortedCurrentValue**
            arr[j], arr[to_swap_ind] = sorted_arr[j], arr[j]
                 
    return swaps              

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input())

    arr = list(map(int, input().rstrip().split()))

    res = minimumSwaps(arr)

    fptr.write(str(res) + '\n')

    fptr.close()
```

## 4) New Year Chaos

It's New Year's Day and everyone's in line for the Wonderland rollercoaster ride! There are a number of people queued up, and each person wears a sticker indicating their initial position in the queue. Initial positions increment by $1$ from $1$ at the front of the line to $n$  at the back.

Any person in the queue can bribe the person directly in front of them to swap positions. If two people swap positions, they still wear the same sticker denoting their original places in line. One person can bribe at most two others. For example, if $n=8$  and $Person=5$ $Person=4$ bribes , the queue will look like this: $1,2,3,5,4,6,7,8$.

Fascinated by this chaotic queue, you decide you must know the minimum number of bribes that took place to get the queue into its current state!


**Task**:

Complete the function minimumBribes in the editor below. It must print an integer representing the minimum number of bribes necessary, or Too chaotic if the line configuration is not possible.


```python
## Most straight-forward solution is to make use of bubble sort

def minimumBribes(q):
    
    swaps = 0
    swaped = False
    
    
    # decrease Q by 1 to make index-matching more intuitive
    # so that our values go from 0 to N-1, just like our
    # indices.  (Not necessary but makes it easier to
    # understand, especially when we substract P-i > 2)
    Q = [P-1 for P in q]
    
    # check if the queue is too chaotic
    for i, v in enumerate(Q):
        if v - i > 2:
            return "Too chaotic"
        
    # bubble sorting to find the answer
    for i in range(0, len(q) - 1):
        for j in range(0, len(q) - 1):
            
            # Bubble sort
            if q[j] > q[j+1]:
                temp = q[j]
                q[j] = q[j+1]
                q[j+1] = temp
                
                # Count swaps
                swaps += 1
                swaped = True
        
        if swaped:
            swaped = False
        else:
            break
    return swaps
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the minimumBribes function below.
def minimumBribes(q):
    bribes = 0 # initialize bribes
    
    # decrease Q by 1 to make index-matching more intuitive
    # so that our values go from 0 to N-1, just like our
    # indices.  (Not necessary but makes it easier to
    # understand, especially when we substract P-i > 2)
    Q = [P-1 for P in q]
    
    # Loop over each person P
    for i,P in enumerate(Q):
        
        # First check if any P is more than two ahead of 
        # its original position
        if P-i >2:
            print("Too chaotic")
            return
        
        # From here on out, we don't care if P has moved
        # forwards, it is better to count how many times
        # P has RECEIVED a bribe, by looking at who is
        # ahead of P.  P's original position is the value
        # of P.
        # Anyone who bribed P cannot get to higher than
        # one position in front if P's original position,
        # so we need to look from one position in front
        # of P's original position to one in front of P's
        # current position, and see how many of those 
        # positions in Q contain a number large than P.
        # In other words we will look from P-1 to i-1,
        # which in Python is range(P-1,i-1+1), or simply
        # range(P-1,i). To make sure we don't try an
        # index less than zero, replace P-1 with
        # max(P-1,0)
        for j in range(max(P-1,0),i):
            if Q[j] > P:
                bribes += 1
    print(bribes)

    
if __name__ == '__main__':
    t = int(input())

    for t_itr in range(t):
        n = int(input())

        q = list(map(int, input().rstrip().split()))

        minimumBribes(q)
```
