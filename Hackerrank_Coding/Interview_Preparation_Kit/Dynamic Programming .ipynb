## 1) Max Array Sum

Given an array of integers, find the subset of non-adjacent elements with the maximum sum. Calculate the sum of that subset.

The solution is taken from [here](http://blog.gainlo.co/index.php/2016/12/02/uber-interview-question-maximum-sum-non-adjacent-elements/) where one can find an explicit explanation of the problem.

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the maxSubsetSum function below.
def maxSubsetSum(arr):
    
    prevOne = 0
    prevTwo = 0
    result = 0
    
    for i in range(len(arr)):
        
        if i == 0:
            result = arr[0]
        
        elif i == 1:
            result = max(arr[0], arr[1])
        
        else:
            result = max(prevOne, prevTwo + arr[i])
        
        prevTwo = prevOne
        prevOne =  result
        
    return result
            

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input())

    arr = list(map(int, input().rstrip().split()))

    res = maxSubsetSum(arr)

    fptr.write(str(res) + '\n')

    fptr.close()
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the maxSubsetSum function below.
def maxSubsetSum(arr):
    
    v = [0] * n
    for i in range(n):
        
        if i < 2:
            v[i] = max(arr[i], 0)
            continue
        elif i == 2:
            v[i] = v[i - 2] + max(arr[i], 0)
            continue

        v[i] = max(v[i-2], v[i-3]) + max(arr[i], 0)

    return max(v[-1], v[-2])
            

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input())

    arr = list(map(int, input().rstrip().split()))

    res = maxSubsetSum(arr)

    fptr.write(str(res) + '\n')

    fptr.close()
```
